```python
# Combines the reviews for test and train data 
# Consolidates IMDB scores for each  review from the name of the file

import os

def CombineReviews():
    
    TestFileName = 'aclImdb/test/testReviews'+".txt"
    if os.path.exists(TestFileName) == True:    
        open(TestFileName, "w").close()  
    
    TrainFileName = 'aclImdb/train/trainReviews'+".txt"
    if os.path.exists(TrainFileName) == True:    
        open(TrainFileName, "w").close()

    Testfile = open(TestFileName, 'a', encoding="utf8")
    Trainfile = open(TrainFileName, 'a', encoding = "utf8")
    
    for i in sorted(os.listdir('aclImdb/test/neg')):    
        with open('aclImdb/test/neg/' + i, encoding="utf8") as file:
            Testfile.write(file.read() + '\n')

    for i in sorted(os.listdir('aclImdb/test/pos')):    
        with open('aclImdb/test/pos/' + i, encoding="utf8") as file:
            Testfile.write(file.read() + '\n')

    
    for i in sorted(os.listdir('aclImdb/train/neg')):    
        with open('aclImdb/train/neg/' + i, encoding="utf8") as file:
            Trainfile.write(file.read() + '\n')

    for i in sorted(os.listdir('aclImdb/train/pos')):    
        with open('aclImdb/train/pos/' + i, encoding="utf8") as file:
            Trainfile.write(file.read() + '\n')
    
    Testfile.close()
    Trainfile.close()

CombineReviews()
    


def IMDBScores():
    TestFileName = 'aclImdb/test/IMDBValuesTest'+".txt"
    if os.path.exists(TestFileName) == True:    
        open(TestFileName, "w").close()  
    
    TrainFileName = 'aclImdb/train/IMDBValuesTrain'+".txt"
    if os.path.exists(TrainFileName) == True:    
        open(TrainFileName, "w").close()

    Testfile = open(TestFileName, 'a', encoding ="utf8")
    Trainfile = open(TrainFileName, 'a', encoding = "utf8")
    
    for i in sorted(os.listdir('aclImdb/test/neg')):
        firstSlice = i.split("_", 1)[1]
        value = firstSlice.split(".", 1)[0]
        Testfile.write(value + '\n')
        
    for i in sorted(os.listdir('aclImdb/test/pos')):
        firstSlice = i.split("_", 1)[1]
        value = firstSlice.split(".", 1)[0]
        Testfile.write(value + '\n')
        
    for i in sorted(os.listdir('aclImdb/train/neg')):
        firstSlice = i.split("_", 1)[1]
        value = firstSlice.split(".", 1)[0]
        Trainfile.write(value + '\n')
        
    for i in sorted(os.listdir('aclImdb/train/pos')):
        firstSlice = i.split("_", 1)[1]
        value = firstSlice.split(".", 1)[0]
        Trainfile.write(value + '\n')
    
    Testfile.close()
    Trainfile.close()
    
    testValues = 'aclImdb/test/testValues'+".txt"
    testV = open(testValues, 'a', encoding = "utf8")
    testV.write('yes')
    testV.close()

#CombineReviews()
IMDBScores()
```
